###  Question 1a
Provide a "misery index" for airports. That is, sort the airports in descending order of the probability that a flight departing from that airport has a delay. Take care of outliers: some airports may have a preposterously low number of flights. We are not interested in them. Your criterion for outliers will be the airports in the lowest 1% percentile in the number of flights.

In the first place we should read the flights dataset.

```python
flight_data_2018 = spark\
  .read\
  .option("inferSchema", "true")\
  .option("header", "true")\
  .csv("D:\\Users\\astar\\Desktop\\spark-5-5-20\\flights.csv\\515364771_T_ONTIME_REPORTING.csv")
```

We will convert `DataFrame` to a table or a view as follows:

```python
flight_data_2018.createOrReplaceTempView("flight_data_2018")
```

Now let's perform a query in SQL in order to see the structure of our dataset. For a more user-friendly output, we will use the  toPandas() action, which collects all rows and returns a pandas DataFrame.
Here is the sample of the first ten rows of our dataset.

```python
structure_query = spark.sql("""
    SELECT *
    FROM flight_data_2018
    LIMIT 10
    """)
structure_query.toPandas()
```

We will count the flights per Origin.

```python
groupByOrigin = spark.sql("""
    SELECT ORIGIN, count(1) as FlightsPerOrigin
    FROM flight_data_2018
    GROUP BY ORIGIN
    order by count(1)
    """)
groupByOrigin.toPandas()
```

At this point we will calculate the lowest 1% percentile in the number of flights which is our criterion for outliers.

```python
threshold = groupByOrigin.toPandas()["FlightsPerOrigin"].quantile(0.01)
```

```python
threshold
```

Based on our results, the number of flights per Origin must be greater than or equal to 60.85 in order not to be consider as an outlier. So, there are 4 outliers - airports and 130 rows will be removed from the dataset:
* 0	YNG	2 
* 1	ART	25
* 2	IFP	45
* 3	CYS	58

```python
new_flight_data_2018 = spark.sql("""
    SELECT *
    FROM flight_data_2018 
    WHERE ORIGIN NOT IN ('YNG', 'ART', 'IFP', 'CYS')
    """)
new_flight_data_2018.createOrReplaceTempView("new_flight_data_2018")
```

Now we are going to check if the 4 airports have been removed succesfully.

```python
new_groupByOrigin = spark.sql("""
     SELECT ORIGIN, count(1) as FlightsPerOrigin
    FROM new_flight_data_2018
    GROUP BY ORIGIN
    ORDER BY count(1)
    """)
new_groupByOrigin.toPandas()
```

We are going to group the number of delayed flights per ORIGIN.

```python
delayed = spark.sql("""
     SELECT ORIGIN, count(1) as DelayedFlightsPerOrigin
    FROM new_flight_data_2018
    WHERE DEP_DELAY > 0
    GROUP BY ORIGIN
    order by count(1)
    """)
delayed.toPandas()
```

We merge the table containing the total flights per Origin with the table of the delayed flights per Origin. 
Then, we calculate the misery index (delayed flights / total flights in descending order) .

```python
new_groupByOrigin.createOrReplaceTempView("new_groupByOrigin")
delayed.createOrReplaceTempView("delayed")
ratio = spark.sql("""
    SELECT new_groupByOrigin.ORIGIN, DelayedFlightsPerOrigin, FlightsPerOrigin, 
            DelayedFlightsPerOrigin/FlightsPerOrigin as misery_index
    FROM new_groupByOrigin, delayed
    WHERE new_groupByOrigin.ORIGIN = delayed.ORIGIN
    ORDER BY misery_index DESC
    """)
ratio.toPandas()
```

### Question 1b

After you have done that, go around your data again, but this time you will show the average and median delay for each airport. You may sort the results by either.

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window
import pyspark.sql.functions as F
median = F.expr('percentile_approx(DEP_DELAY, 0.5)')
mean_median_dat = new_flight_data_2018\
        .groupBy("ORIGIN")\
        .agg(mean("DEP_DELAY").alias('meanOfDepartDelay'), median.alias('medianOfDepartDelay'))\
        .sort(asc("meanOfDepartDelay"))
mean_median_dat.toPandas()
```

### Question 1c
Finally, enrich your airport misery index by tabulating both the probability, in descending order, that you will experience a delay at a given airport, and the average and median delay that you are likely to experience.

```python
ratio.createOrReplaceTempView("ratio")
mean_median_dat.createOrReplaceTempView("mean_median_dat")
mis_mean_med = spark.sql("""
    SELECT ratio.ORIGIN,misery_index, meanOfDepartDelay, medianOfDepartDelay      
    FROM ratio, mean_median_dat
    WHERE ratio.ORIGIN = mean_median_dat.ORIGIN
    ORDER BY ratio.misery_index DESC
    """)
mis_mean_med.toPandas()
```

### Question 2a
Provide a "misery index" for airlines. That is, do the same thing you did for the airports, but this time we are interested in the airlines that make life difficult for passengers. Sort the airlines in descending order of probability that a flight operated by the airline has a delay. This time we do not care about outliers.

We will count the flights per carrier.

```python
groupByCarrier = spark.sql("""
     SELECT CARRIER, count(1) as FlightsPerCarrier
    FROM flight_data_2018
    GROUP BY CARRIER
    ORDER BY count(1)
    """)
groupByCarrier.toPandas()
```

We are going to group the number of delayed flights per carrier.

```python
delayedCarrier = spark.sql("""
     SELECT CARRIER, count(1) as DelayedFlightsPerCarrier
    FROM flight_data_2018
    WHERE DEP_DELAY > 0
    GROUP BY CARRIER
    order by count(1)
    """)
delayedCarrier.toPandas()
```

We merge the table containing the total flights per carrier with the table of the delayed flights per carrier. 
Then, we calculate the misery index (delayed flights / total flights in descending order) .

```python
groupByCarrier.createOrReplaceTempView("groupByCarrier")
delayedCarrier.createOrReplaceTempView("delayedCarrier")
ratio_carrier = spark.sql("""
    SELECT groupByCarrier.CARRIER, DelayedFlightsPerCarrier, FlightsPerCarrier, 
            DelayedFlightsPerCarrier/FlightsPerCarrier as misery_index
    FROM groupByCarrier, delayedCarrier
    WHERE groupByCarrier.CARRIER = delayedCarrier.CARRIER
    ORDER BY misery_index DESC
    """)
ratio_carrier.toPandas()
```
